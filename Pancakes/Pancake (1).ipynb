```
options = {
    "node_color": "white",
    "edgecolors": "black"}
def flip(lis, pos): 
  if pos == 1:
    return lis
  elif pos == len(lis):
    l = lis[::-1]
    return l
  else:
    l1 = lis[0:pos]
    l1 = l1[::-1]
    l2 = lis[pos:]
    return l1 + l2
def GraficaRedPancakes(n): 
  num_permuta = m.factorial(n)
  permutacion = list(range(1,n+1))
  permutaciones = list(itertools.permutations(permutacion))
  G=nx.Graph()
  for i in range(0,num_permuta):
    G.add_node(lstr(permutaciones[i]),color="red")
  for i in range(0, num_permuta):
    for j in range(2, n+1):
      flipped = flip(permutaciones[i], j)
      for k in range(0, num_permuta):
        if flipped == permutaciones[k]:
          edge =  (lstr(permutaciones[i]), lstr(permutaciones[k]))
          G.add_edge(*edge,color="black")
          break
  nx.draw(G, with_labels = True,**options)
  plt.show()
```

```
print("Red Pancake para n = 2")
GraficaRedPancakes(2)
print("Red Pancake para n = 3")
GraficaRedPancakes(3)
print("Red Pancake para n = 4")
GraficaRedPancakes(4)
print("Red Pancake para n = 5")
GraficaRedPancakes(5)
print("Red Pancake para n = 6")
GraficaRedPancakes(6)
```

```
def goodLevel(s,L):  
        if s == sorted(s):
         return True
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
         
        queue = [] 
  
        queue.append([s,0]) 
        visited[lstr(s)] = True
  
        while queue: 
  
            e = queue.pop(0)
            s = e[0]
            l = e[1] 
            graph[lstr(s)]=[]

            flag = False
            for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
                if(d == sorted(d)):
                     flag = True
            if flag:
              node = pydot.Node(lstr(s),style="filled", fillcolor="yellow")
            else: 
              node = pydot.Node(lstr(s))

            G.add_node(node)
            
            if(s != sorted(s)) and (l < L):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                     node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") 
                    return True
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append([d,l+1]) 
                  visited[lstr(d)] = True 
              
        return False
def Numberlittle(n):
  if n == 1:
    return 0
  max = 1
  num_permuta = m.factorial(n)
  permutacion = list(range(1,n+1))
  permutaciones = list(itertools.permutations(permutacion))
  for i in range(0,num_permuta):
    for j in range(max, 2*n):
      perm = list(permutaciones[i])
      if goodLevel(perm, j):
        max = j
        break
  return max
```

```
def goodLevels(s,L):  
        if s == sorted(s):
         return True
        graph = defaultdict()
        visited = defaultdict()
        queue = [] 
        queue.append([s,0]) 
        visited[lstr(s)] = True
  
        while queue: 
            e = queue.pop(0)
            s = e[0]
            l = e[1] 
            graph[lstr(s)]=[]
            if(s != sorted(s)) and (l < L):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                  if(d == sorted(d)):
                    return True
                  graph[lstr(s)].append(lstr(d))
                  queue.append([d,l+1]) 
                  visited[lstr(d)] = True 
        return False
def simplifiedPancake(n):
  if n == 1:
    return 0
  max = 1
  num_permuta = m.factorial(n)
  permutacion = list(range(1,n+1))
  permutaciones = list(itertools.permutations(permutacion))
  for i in range(0,num_permuta):
    for j in range(max, 2*n):
      perm = list(permutaciones[i])
      if goodLevels(perm, j):
        max = j
        break
  return max
```

```
def twoHandFlip(lis, pos1, pos2): 
  l1= []
  l2 = []
  l3 = []
  if pos1 == pos2:
    return lis
  if pos1 > 1:
    l1 = lis[0:pos1-1]
  if pos2 < len(lis):
    l3 = lis[pos2:]
  l2 = lis[pos1-1:pos2]
  l2 = l2[::-1]
  return l1 + l2 + l3
```
